```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os


csv_data = pd.read_csv("format1.csv")
excel_data = pd.read_excel("format2.xlsx")
json_data = pd.read_json("format3.json")

csv_data


csv_data.head()


excel_data.head()


json_data.head()

#merging dataset 

df = pd.concat([csv_data, excel_data, json_data])
df


df.head()

df.describe()

df.info()


#Data Transformation


df.Date = pd.to_datetime(df.Date)
df.Time = pd.to_datetime(df.Time)

df.head()

df.dtypes


## Data cleaning

df.isna().sum()

df.duplicated().sum()



# Data analysis

total_sales = df['Total'].sum()
total_sales

average_order_value = df.groupby('Gender')['Total'].mean()
average_order_value


product_category_distribution = df['Product line'].value_counts()
product_category_distribution



df.columns


## Data visualization

df.boxplot()

plt.figure(figsize=(12, 6))
plt.bar(df['Product line'], df['Total'], color='skyblue')
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales by Product Category')
plt.xticks(rotation=20)
plt.show()
```
